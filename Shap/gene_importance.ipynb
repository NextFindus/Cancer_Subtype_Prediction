```python
import pandas as pd
import numpy as np
```

```python
tissue_types=['All','Saliva','Blood','Brain','Buccal','Skin']
manifest=pd.read_csv('/sybig/home/fim/Dokumente/Sysbio/All_models/SHAP/humanmethylation450_15017482_v1-2.csv', engine='python', on_bad_lines='skip',skiprows=7)
print(manifest)
print(list(manifest.columns))
```

```python
shap=np.load("/sybig/home/fim/Dokumente/Sysbio/All_models/SHAP/shap_values__AltumAge_80:20_All.npy", allow_pickle=True)
print(shap.head)
```

```python


merged_df = pd.merge(shap, manifest, left_on='CpG_ID', right_on='CpG_ID')

# Step 3: Calculate mean SHAP value per gene
mean_shap_per_gene = merged_df.groupby('Gene_ID')['SHAP_value'].mean().reset_index()

# Step 4: Find gene with highest mean SHAP value
top_gene = mean_shap_per_gene.loc[mean_shap_per_gene['SHAP_value'].idxmax()]

print(f"Gene with highest mean SHAP value: {top_gene['Gene_ID']}")
print(f"Mean SHAP value: {top_gene['SHAP_value']}")

# Optional: Save the mean SHAP per gene for further analysis
mean_shap_per_gene.to_csv('mean_shap_per_gene.csv', index=False)
```
